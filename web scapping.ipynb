```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.flipkart.com/search?q=iphone%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
allow = requests.get(url)
```

```python
allow
```

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

```python
iphone_type = soup.select("._4rR01T")
```

```python
iphone_type
```

```python
iphone_list = []
product_junk = []
```

```python
for item in iphone_type:
    iphone_list.append(item.text)
    junk = "https://www.flipkart.com"
    product_junk.append(junk)
```

```python
iphone_list
```

```python
iphone_cost= soup.select("._1_WHN1")
```

```python
iphone_cost
```

```python
rate_list = []
```

```python
for item in iphone_cost:
    rate_list.append(item.text)
```

```python
rate_list
```

```python
import pandas as pd
```

```python
iphone = pd.DataFrame(iphone_type,columns=["iphone Model"])
```

```python
iphone['Rate']=rate_list
```

```python
iphone
```
